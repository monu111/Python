```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_values = [1400,600,300,410,250]
exp_label = ['Home rent','food','phone/internet','car','other']
```

```python
plt.pie(exp_values,labels = exp_label)
```

```python
plt.axis('equal')
plt.pie(exp_values,labels = exp_label)
plt.show()  # then we can remove unwanted no. which is given in the upperside 
```

```python
# we can increse the size of the pie chart using the radius 
plt.axis("equal")
plt.pie(exp_values,labels = exp_label,radius = 2) # by default radius 1& rad = 2 increse the size of pie chart
plt.show()
```

### autopct = enable to display the percentage using the python string formatting

```python
plt.axis('equal')
plt.pie(exp_values, labels = exp_label,radius = 1.5,autopct = '%0.1f%%') # 0.0 = 0 desimal point,0.2 = 2 floting no.
plt.show()
```

###  EXPLODING PIE CHART

```python
plt.axis('equal')
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%.2f',shadow = True)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels = exp_label,radius = 1.5, autopct = '%0.2f%%', shadow = True, explode = [0,0.3,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0.4,0,0])
plt.show()
```

```python
plt.axis = "equal"
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0.2,0.2,0.1])
plt.show()
```

# STARTANGLE = set pie chart with diff. angle, 

```python
plt.axis = "equal"
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0],startangle = 45)
plt.show()
```

```python
plt.axis = "equal"
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0],startangle = 180)
plt.show()
```

```python
plt.axis = "equal"
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0],startangle = 80)
plt.show()
```

```python
# how to save piechart in a file
plt.axis = ("equal")
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0])
#plt.show()
plt.savefig('piechart.png')
```

```python
# how to save piechart in a file & if chart is not well fitting then we use bbox_inche
plt.axis = ("equal")
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0])
#plt.show()
plt.savefig('piechart.png')
```

```python
# how to save piechart in a file
plt.axis = ("equal")
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0])

plt.savefig('piech.png',bbox_inches = 'tight')
```

```python
# how to save piechart in a file
plt.axis = ("equal")
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0])
#plt.show()
plt.savefig('piechart.png',bbox_inches = 'tight',pad_inches = 2,transparent = True)
```

```python
# save pie chart with diff. directries
# how to save piechart in a file
plt.axis = ("equal")
plt.pie(exp_values,labels = exp_label,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0])
#plt.show()
plt.savefig('D:\\piechart.pdf',bbox_inches = 'tight',pad_inches = 2,transparent = True)
```
