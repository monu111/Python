```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color = 'green',linewidth = 5,linestyle ='dotted' )
```

```python
plt.plot(x,y,color  = 'green',linewidth = 5,linestyle = 'dashdot')
```

```python
plt.plot(x,y,color = 'red',linewidth = 5,linestyle = 'dotted',)
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title("weather")
plt.plot(x,y,color = 'yellow',linewidth = 3,linestyle = 'dashdot',)
```

```python
#format string used in 2d plot  ,any kind of chart
plt.plot(x,y,'g+') # specfied  format string as 3rd argument
```

```python
plt.plot(x,y,'g+--')
```

```python
# 
plt.plot(x,y,'--r+')
```

```python
plt.plot(x,y,'rD') # d = diamond(marker)
```

```python
# WE can also print in this format
plt.plot(x,y,color = 'red',marker = 'D',linestyle = 'dotted')
```

```python
#NOTE = format string work as 3 argument color,marker,linestyle, alpha control the transprency of the line
# value of alpha  b/w  0 to 1 
```

```python
plt.plot(x,y,color = 'red',marker = '+',linewidth = 5,linestyle = '-',markersize = '20',alpha = 0.5)
```

# AXIS LABEL,legend,grid

```python

max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,38,48,42,41,42,43]
days = [1,2,3,4,5,6,7]
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')

```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')

plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(days,max_t,label = 'Max')
plt.plot(days,min_t,label = 'Min')
plt.plot(days,avg_t,label = 'Avg')
plt.legend()
```

```python
plt.xlabel('days')
plt.ylabel("temperature")
plt.title('weather')
plt.plot(days,max_t,label = 'Max')
plt.plot(days,min_t,label = 'Min')
plt.plot(days,avg_t,label = 'Avg')
plt.legend(loc = 'upper right')
```

```python
# fontsize = large,small,shadow on the legend
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(days,max_t,label = 'Max')
plt.plot(days,min_t,label = 'min')
plt.plot(days,avg_t,label = 'avg')
plt.legend(loc = 'best',shadow = True,fontsize = 'small')
```

```python
# fontsize = large,small,shadow on the legend
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(days,max_t,label = 'Max')
plt.plot(days,min_t,label = 'min')
plt.plot(days,avg_t,label = 'avg')
plt.legend(loc = 'best',shadow = True,fontsize = 'small')
plt.grid()
```
