# Violinplot

A violin plot is a method of plotting numeric data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side.

Violin plots are similar to box plots, except that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator.

- combo of boxplot+ kernel density estimation.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python

sns.violinplot(x = 'sex', y = 'total_bill', data = df)
```

```python
sns.violinplot(x = 'total_bill', y = 'sex', data = df)
```

```python
plt.figure(figsize = (20,20))
sns.violinplot(x = 'sex', y = 'total_bill', data = df)
```

```python
 sns.violinplot(x="day", y="total_bill", hue="sex",
                    data=df, palette="Set2", split=True,
                        scale="count")
```

- it is combo kde and box plot.
- inside the violinplot there is box plot.
- it also shows the  probability density . we can see in the above graph.
