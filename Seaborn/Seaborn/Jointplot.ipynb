# Seaborn

Different types of analysis:
- Univariate (U) : In univariate analysis we use a single feature to analyze its properties.
    
- Bivariate (B): When we compare the data between exactly 2 features then its called bivariate analysis.
    
- Multivariate (M): Comparing more than 2 variables is called as Multivariate analysis.

## JointPlot

- A joint plot allow us to study the relationship b/w 2 numeric variables. The central chart display their correlation.

- it is usually a scatterplot, a hexbin plot, a 2D density plot.

- Use for bivariate analysis.

Join plots can do both univariate as well as bivariate analysis. The main plot will give us a bivariate analysis, whereas on the top and right side we will get univariate plots of both the variables that were considered. It makes our job easy by getting both scatter plots for bivariate and Distribution plot for univariate, both in a single plot.

```python
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset.head()
```

```python
dataset.dtypes
```

```python
sns.jointplot(x = 'EstimatedSalary', y = 'Purchased', data = dataset) # by default kind = 'scatter'
```

```python
# Using kind  = hex
```

```python
sns.jointplot(x = 'EstimatedSalary', y = 'Purchased', data = dataset, kind = 'scatter')
```

```python
sns.jointplot(x = ['EstimatedSalary'], y = ['Purchased'], data = dataset, kind = 'hex')
```

```python
sns.jointplot(x = 'EstimatedSalary', y = 'Purchased', data = dataset, kind = 'kde')
```

- in the above graph Kernel Density estimation(kde)

```python
sns.jointplot(x = 'EstimatedSalary', y = 'Purchased', data = dataset, kind = 'reg') 
```
